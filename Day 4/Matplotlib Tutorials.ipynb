# Matplotlib Tutorials

The purpose of this tutorial is to introduce you to some of the core ploting functions and methods in the Matplotlib python library. You will be required to do some more reading around Matplotlib to learn about the rest of the plotting functions, methods and its properties.

## Import Relevant Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Getting Started

When running python in the command line, graphs are displayed in the a separate window. However, in Jupyter Notebook, you can use the `%matplotlib inline` magic command to output graphs within the notebook

In matplotlib, you can configure the graph format to a number of `InlineBackend.figure_format` to give your graphs different look and feel as preferred. Examples of those are **svg** and **retina**

For this tutorial, we will set the `InlineBackend.figure_format` to `retina`

```python
%config InlineBackend.figure_format = 'retina'
```

Additionally, matplotlib comes with an array of plotting styles which you can choose from. For the purposes of this tutorial we will use `seaborn`. To see a list of all the style plots, click [here](https://matplotlib.org/3.2.1/gallery/style_sheets/style_sheets_reference.html)

```python
plt.style.use("seaborn")
```

The Matplolib `plot` function at a basic level takes in a list-like object in order to plot. In the event that only one list-like object is provided, it will use the index of the items in the list as the x-values and the items themselves as the y-values. 

```python
a_list = [4,2,9,7,4,6]
plt.plot(a_list)
plt.show()
```

In the event that two lists are provided, it will use the first list as the x-values and the second list as the y-values.

```python
c_list = [3,5,6,7,8,9]
b_list = [23,56,78,65,70,88]
plt.plot(c_list, b_list)
plt.show()
```

The `plot` function automatically detects the range of the x and y values and draws the plot to match the ranges. However, we can also specify manually the range of the x and y values using the `xlim` and `ylim` methods.

```python
plt.plot(c_list, b_list)
plt.xlim(0,10)
plt.ylim(15,100)
plt.show()
```

Next, we are going to play around with some of the parameters in the `plot` function and some additional methods. We will use numpy's `linspace` function to create an array `x` containing 1000 floats from -5 to 5. We will then make the y = x**2

One of the parameters in the `plot` function is the `color`. You can set the `color` to different colors as preferred. You may refer to the Matplotlib documentation for a list of all available colors [here](https://matplotlib.org/2.0.2/api/colors_api.html).

```python
x = np.linspace(-5,5,1000)
y = x**2
plt.plot(x,y, color='blue')
plt.show();
```

Let's add some text to the plot to give some more information

```python
plt.plot(x,y, color='green')
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.show()
```

## Line and Style Color

By default, matplotlib draws a line to join two points in the graph. You may manually specify the line style of your choice. You may also add some colors to the line styles as well. See [here](https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html) to read more.

In the graph below, used `g--` which means a green dashed line style. You can play around with different line styles and colors

```python
plt.plot(x,y, 'g-.')
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.show()
```

You can also plot multiple lines on the same graph. You can to use the `plot` function a number of times before calling the `show()`

```python
plt.plot(x,y, 'g-')
plt.plot(x,x**3, 'b:')
plt.plot(x,x**4, 'r-.')
plt.title("Multiple Line Plots")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.show()
```

## Bar Plot

Bar plot is used to plot categorical values. It takes in categorical values as x and its corresponding y values and height.

```python
course_names = ['Eshete','Nyagaka','Mwango', 'Mansa', 'Kofi']
scores = [90,98,79,69,89]
plt.bar(course_names, scores)
plt.title("A Bar plot of students and their scores")
plt.xlabel("Students in a Couse")
plt.ylabel("Scores")
plt.show;
```

You can also compare two or more values across categorical values. For example, you can compare the performance of males and females across year groups.

```python
male_scores = [87,98,56,99]
female_scores = [98,88,55,67]
ind = np.arange(len(male_scores)) # creates a numpy array from 0 to 3. it will be used on the x-axis
width = 0.4 # will be used to specify the width of the bar
plt.xticks(ind + width / 2, ('Freshmen', 'Sophomores', 'Juniors', 'Seniors'))
plt.bar(ind, male_scores, width, label="Males")
plt.bar(ind + width, female_scores, width, label="Females")
plt.title("A Bar Plot comparing the performance of males and females across four classes in a University")
plt.xlabel("Class")
plt.ylabel("Scores")
plt.legend()
plt.show()
```

## Histogram

```python
import random # this will help us to generate some random numbers
```

In the plot below, I generated 200 random numbers between 150 and 200 and stored it in `x_values`. I then used it in the `plt.hist` to plot the histogram to show the distribution of the random integers.

```python
x_values1 = []
x_values2 = []
x_values3 = []
for i in range(200):
    x_values1.append(random.randint(150,200))
    x_values2.append(random.randint(150,200)+25)
    x_values3.append(random.randint(150,200)+75)
```

```python
plt.hist(x_values1, bins=20)
plt.title("A Histogram showing the distribution of 200 randomly generated integers between 150 and 200")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show;
```

We can also plot multiple `x_values` on the same graph. This can be useful when you want to compare two or more things.

```python
plt.hist(x_values1, bins=10, label='bar hist')
plt.hist(x_values2, histtype='stepfilled',bins=10, label='stepfilled')
plt.hist(x_values3, bins=10, label='bar hist')
plt.legend()
plt.show();
```

## Boxplot

You can plot a boxplot using the `boxplot` function. There are a lot of other parameters to be explored. Do well to learn more about Boxplot plotting in Matplotlib.

```python
plt.boxplot(x_values1, vert=False, showmeans=True)
plt.title("A Boxplot showing of 200 randomly generated integers between 150 and 200")
plt.show();
```

## Scatter Plot

```python
plt.scatter(x_values1, x_values3)
plt.title("A scatter plot showing the relationship between two lists of randomly generated integers between 150 and 275")
plt.show()
```

You may also add scale, `s` to the plot to have different sizes for each of the points

```python
scale = np.random.rand(200)
scale = 100 * scale ** 2
plt.scatter(x_values1, x_values3, s=scale)
plt.title("A scatter plot showing the relationship between two lists of randomly generated integers between 150 and 275")
plt.show()
```

You can also color your points as preferred

```python
for i in ['red', 'blue', 'green']:
    n = 20
    x,y = np.random.rand(2,n)
    scale = 300 * np.random.rand(n) ** 2
    plt.scatter(x,y, s=scale, c=i, alpha=0.7)
plt.show()
```

## Text and Annotations

Matplotlib has a `text()` function that allows you to write text to any part of the graph. All you need to do is to specify the coordinates and other optional parameters.

```python
x = np.linspace(-2,2, 500)
y = x**2
px = 0.9 # the x coordinate of the position we want to place our text
py = px **2 # the y coordinate of the position we want to place our text
plt.plot(x,y, 'r-', px, py, 'bo') # the px and py indicate where we want to put the dot
plt.text(0,3.0,'Quadratic Function\n$y = x^2$', fontsize=17, color='green', horizontalalignment='center')
plt.text(px+0.09,py-0.4, 'x = %0.2f\ny = %0.2f'%(px,py), rotation=50, color='gray')
plt.show();
```

## Subplots

Subplots allows us to plot multiple graphs on a grid. If you want to create subplots, use the `subplot()` method specifying the number of rows and columns, and the index of the subplot you want to draw on. It starts on 1 - top left to top right and then bottom left to bottom right.

```python
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(12, 6))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(x, x**(i+1))
plt.show();
plt.savefig('subplot.jpg',transparent=True)
```

## Saving a Plot

To save a plot, all you have to do is all the `savefig()` method and name the plot you want to save and add the extension.

```python
x = np.linspace(-9,9,3000)
plt.plot(x,x**2)
plt.savefig('quad_plot.jpg',transparent=True)
```
