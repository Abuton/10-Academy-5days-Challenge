#### Question 1. Make a scatter plot of the Longitude and Latitude columns

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set()

df = pd.read_csv('data/pandas_tasks.csv')
df.head()
```

```python
plt.figure(figsize=(12,7))
ll = sns.scatterplot(x='Latitude', y='Longitude', data=df, color='g')
ll.set(xlabel='Latitude', ylabel='Longitude', title='Latitude Vs Longitude')
plt.show()
```

#### Question 2. Make a scatter plot of the Longitude and Latitude columns. This time, plot with 'hue' = Q2 and compare it over Q3

```python
plt.figure(figsize=(12,7))
ll_hue = sns.scatterplot(x='Latitude', y='Longitude', data=df, color='g', hue='Q2')
ll_hue.set(xlabel='Latitude', ylabel='Longitude', title='Longitude Vs Latitude with  Hue Q2')
plt.show()
```

```python
plt.figure(figsize=(12,7))
ll_hue = sns.scatterplot(x='Latitude', y='Longitude', data=df, hue='Q3')
ll_hue.set(xlabel='Latitude', ylabel='Longitude', title='Longitude Vs Latitude with  Hue Q3')
plt.show()
```

#### Question 3. Make a boxplot of Longiture and Q2

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Q2', y='Longitude', data=df)
plt.title('Box Plot of Longitude as spread across Q2')
plt.show()
```

#### Question 4. Make a count plot of all the cateegorical variables in the dataset.

```python
# since all the columns has been encoded in a numerical way hence i used intuition to select category columns
# as those with a contable diffrent values i.e unique values. Hence columns like Longitude, Latitude, Age
# are dropped and i plot the rest.

category_columns = df.select_dtypes(include=np.int64)
category_columns = category_columns.drop(['ID', 'Q1'], axis=1)

def plot_counts(data, column):
    plt.figure(figsize=(9, 5))
    count = sns.countplot(x=column, data=data)
    count.set(title=f'Counts of {column}')
    plt.show()
```

```python
for column in category_columns.columns:
    plot_counts(df, column)
```
