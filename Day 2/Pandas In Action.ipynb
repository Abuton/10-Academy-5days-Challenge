## Series and DataFrames

There are two types of data structures in Pandas: **Series** and **DataFrame**.

**Series** - It is similar to a python list. Thus, it is a one dimensional data structure that can store values and you can use indexing to access values in the Series object. You can initialize a Series object with a list as shown below.

```python
import pandas as pd
series = pd.Series([67,34,16,29,69,20])
print(series)
```

```python
series.
```

The Series object in pandas come with numerous functions and methods. You can use the code completion feature in Jupyter notebook to see the list of all the available functions and methods for the Series object. Type "series." in a cell and hit the Tab key on your keyboard.

**DataFrame** - This is like a normal excel spreadsheet which has rows and columns. Pandas has the **pd.DataFrame()** for creating a dataframe. There are a number ways of creating a dataframe. We shall go through each one of them.

In first example below, I created a DataFrame object using the **pd.DataFrame()** object. In the dataframe object, I entered a list of items which are supposed to represent the rows. Remember that if you want to enter multiple rows, you have to create multiple lists in one main list that will contain all the other lists (representing the rows).
In addition, I indicated a **columns** parameter and gave it a number of column names (Name, age, country and University). Ensure that each item in the list corresponds to each item in the columns.

```python
df1 = pd.DataFrame([["Sebastian", "24", "Ghana", "Ashesi University"],["Ama", "23", "Togo", "Togo University"],
                   ["Hannah", "20", "Congo", "Congo University"],["John", "21", "Benin", "Benin University"],
                  ["Gloria", "28", "Morocco", "Hoguan University"],["Ben", "25", "Kenya", "Kenyatta University"]],
    columns=["Name", "Age", "Country", "University"])

df1
```

Another way of creating a DataFrame object is using dictionaries. Example below

```python
data = {'Name': ['Sebastian', 'Ama', 'Hannah', 'John', 'Gloria', 'Ben'],
       'Age': [24,23,20,21,28,25],
       'Country': ['Ghana', 'Togo', 'Congo', 'Benin', 'Morocco', 'Kenya'],
       'Univeristy': ['Ashesi University', 'Togo University', 'Congo University','Benin University', 
                      'Hoguan University', 'Kenyatta University']}

df2 = pd.DataFrame(data)
df2
```

Alternatively, you can create an empty dataframe with columns and append the rows to it as dictionaries with column names as keys.

```python
df3 = pd.DataFrame(columns=["Name", "Age", "Country", "University"])
df3 = df3.append({"Name": "Sebastian", "Age": 24, "Country": "Ghana", "University": "Ashesi University"}, ignore_index=True)
df3 = df3.append({"Name": "Hannah", "Age": 20, "Country": "Togo", "University": "Togo University"}, ignore_index=True)
df3 = df3.append({"Name": "John", "Age": 21, "Country": "Congo", "University": "Congo University"}, ignore_index=True)
df3 = df3.append({"Name": "Gloria", "Age": 28, "Country": "Benin", "University": "Benin University"}, ignore_index=True)
df3 = df3.append({"Name": "Ben", "Age": 25, "Country": "Kenya", "University": "Kenyatta University"}, ignore_index=True)
df3 = df3.append({"Name": "Ama", "Age": 23, "Country": "Morocco", "University": "Hoguan University"}, ignore_index=True)

df3
```

Another alternative is to append all rows as lists at the end of the dataframe. In this example, we used the loc property in pandas. The **loc** in pandas is used to access a group of rows columns by lable(s) or a boolean array. In this example, the length of the **df4** is 4. On the **loc** property, we put this length as a label of the index. Please note that contrary to python slices, both the start and end are inclusive. This means for each column of the **df4** insert each item of the list provided on the right side of the equal sign.

```python
df4 = pd.DataFrame(columns=["Name", "Age", "Country", "University"])
df4.loc[len(df4)] = ["Sebastian", "24", "Ghana", "Ashesi University"]
df4.loc[len(df4)] = ["Ama", "23", "Togo", "Togo University"]
df4.loc[len(df4)] = ["Hannah", "20", "Congo", "Congo University"]
df4.loc[len(df4)] = ["John", "21", "Benin", "Benin University"]
df4.loc[len(df4)] = ["Gloria", "28", "Morocco", "Hoguan University"]
df4.loc[len(df4)] =  ["Ben", "25", "Kenya", "Kenyatta University"]
df4
```

## Data Manipulation

### Handling missing values

Handling missing values can be very tiring yet very important in the data science process. We must idetify them first and then understand why they are missing. There are two ways to handle missing values:
1. Remove rows and columns with missing values
2. impute missing values, replacing them with predefined values

Missing values in python are denoted by **'na'** or **'null'** in pandas (both meaning the same thing)
Numpy however denotes missing values as **NaN** meaning Not a Number.

First, lets count the number of missing values in each column

```python
credit_score.isna().sum()
```

Lets check the shape of the dataset before we drop null values

```python
credit_score.shape
```

The function **dropna()** drops rows with at least one missing value.

```python
credit_score_na = credit_score.dropna()
credit_score_na.shape
```

```python
credit_score_na.isna().sum()
```

Lets take a look at **own_telephone** column

```python
credit_score['own_telephone'].value_counts()
```

```python
credit_score['own_telephone'].isna().sum()
```

We can see that the **own_telephone** column has four missing values. We will use the **fillna()** method to replace the missing values with **"unknown"**

```python
credit_score['own_telephone'].fillna("Unknown", inplace = True)

#This way also works
#credit_score[['own_telephone']] == credit_score[['own_telephone']].fillna('Unknown')_

credit_score['own_telephone'].value_counts()
```

To confirm there are no missing value in the **own_telephone** column, we do the following

```python
credit_score.isna().sum()
```

## Handling Irregular Cardinality


The cardinality is the different number of features we have to a particular feature. In a column, you may have different values which represent the a particular feature. When this happens, it is called **Irregular Cardinality**

Lets count the number of unique elements in each column.

```python
credit_score.nunique()
```

When we look at the **credit_history** column, we can see five distinct values in that column. However, some of the unique values can be replaced with another. For example, the **"no credits/all paid"**, **"all paid** and **existing paid** means there is no obligation to make any payment since in each of the mentioned values, moneys have been paid.

```python
credit_score['credit_history'].value_counts()
```

Next, we will use the **replace()** method to replace all problematic levels to a standard set of levels.

```python
credit_score.replace(["'all paid'", "'no credits/all paid'", "'existing paid'"], "All paid", inplace=True)
```

Now lets check all the unique values in the **credit_history** column

```python
credit_score['credit_history'].value_counts()
```

## Detecting Outliers

To use the boxplot to detect outliers, lets use pandas' plotting functions, **Matplotlib**.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
plt.style.use("ggplot")
```
